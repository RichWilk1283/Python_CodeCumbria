I have choosen to do my homework on UFC figthers statistics.


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

ufc_data = pd.read_csv('datasets\\ufc-fighters-statistics.csv', sep=',')
```

```python
ufc_data.head()
```

<h5>Q1:</h5>

Which figthers has the most wins.

```python
ufc_data.sort_values(by=['wins'], ascending=False)[0:1]
```

<h5>Q2:</h5>

Which figther has the worst losses ratio, compared to their wins.

```python
ufc_data['total_fights'] = ufc_data['losses'] + ufc_data['wins'] + ufc_data['draws']
ufc_data['percentage_losses'] = (100 / ufc_data['total_fights']) * ufc_data['losses']

newList = ufc_data[ufc_data['percentage_losses'] == 100]
newList.sort_values(by=['total_fights'], ascending=False)[0:1]
```

<h5>Q3:</h5>

Does a fighters weight class effect their significant strikes per minute stats.

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(18, 8))

fig.suptitle('Weigth vs strikes per minute')

axes.scatter(x=ufc_data['weight_in_kg'], y=ufc_data['significant_strikes_landed_per_minute'], s=5)

axes.set_ylabel('Strikes per minute')
axes.set_xlabel('Weight')
```

<h5>Q4:</h5>

Is there any connection to a fighters reach and thier losses.

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(18, 8))

fig.suptitle('Reach vs losses %')

axes.scatter(x=ufc_data['reach_in_cm'], y=ufc_data['percentage_losses'], s=9)

axes.set_ylabel('% Losses')
axes.set_xlabel('Reach(CM)')
```
